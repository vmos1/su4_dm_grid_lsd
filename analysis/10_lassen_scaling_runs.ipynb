```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import itertools
```

```python
%matplotlib notebook
```

```python
Nc=[4,4,3,3]
Vol=['16_8','32_16','16_8','32_16']
rank1=[180,2770,140,1605]
rank2=[120,1440,105,820]
rank4=[93,796,90,475]
rank8=[75,490,78,330]
```

```python

cols=['nc','vol','rank1','rank2','rank4','rank8']
df=pd.DataFrame(zip(Nc,Vol,rank1,rank2,rank4,rank8),columns=cols)
df
```

```python
cls=[i for i in df.columns if i.startswith('rank')]
```

```python
df.loc[1][cls]/df.loc[0][cls]
```

```python
df.loc[3][cls]/df.loc[2][cls]
```

```python
df.loc[0][cls]/df.loc[2][cls]
```

### Plots

```python
# Plot all 

plt.figure()
x=[1,2,4,8]

for (i,j),marker in zip(df.iterrows(),itertools.cycle('>^*sDHPdpx_')):
    y=j[['rank1','rank2','rank4','rank8']].values
    label='Nc='+str(j['nc'])+'_'+j.vol
    print(y,label)
    plt.plot(x,y,label=label,marker=marker)
    
plt.ylabel('Time per trajectory',rotation='vertical')
plt.xlabel('# of GPUs')
plt.legend()
# plt.show()
plt.savefig('scaling.pdf')
```
